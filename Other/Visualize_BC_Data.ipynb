## Look at pair plots of some variables

We have no idea what the variables are, so let's literally copy the first five from above and look at them.

```python
sub_cols = ['mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity']
sns.pairplot(bc_df, hue = 'target_category', vars = sub_cols)
```

"mean perimeter" looks kind of diagnostic...

Let's look at the next five:

```python
sub_cols = ['mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error']
sns.pairplot(bc_df, hue = 'target_category', vars = sub_cols)
```

---

It looks like "mean concave points" here and "mean perimeter" from above do pretty well. Perhaps we should use them. Let's look at them together!

```python
sub_cols = ['mean concave points', 'mean perimeter']
sns.pairplot(bc_df, hue = 'target_category', vars = sub_cols)
```

This looks promising. Perhaps we should pursue these 2 variables.
