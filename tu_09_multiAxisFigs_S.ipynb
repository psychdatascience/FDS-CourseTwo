```python
# first things first
import numpy as np
import matplotlib.pyplot as plt
```

```python
### make some data to play with
my_means = [0, 0]
my_cov = [[2, -1.9], [-1.9, 3]]
my_n = 5000

my_rng = np.random.default_rng()
x, y = my_rng.multivariate_normal(my_means, my_cov, my_n).T
y = y**2
```

```python
np.corrcoef(x,y)
```

```python

fig = plt.figure(figsize = (10,10))                        # make the figure and name it fig
fig.set_facecolor('gray')                 # set the face color to gray

# y subplot
yhist_ax = plt.subplot(2,2,1)             # name our y histogram axes object
plt.hist(y, orientation = 'horizontal');
plt.yticks([])
yhist_ax.invert_xaxis()                   # have the axes invert its x axis
plt.title('distribution of y');

# x subplot()
plt.subplot(2,2,4)
plt.hist(x);
plt.xticks([])
plt.title('distribution of x');

# scatterplot subplot
plt.subplot(2,2,2)
plt.scatter(x,y, alpha = 0.1);
plt.xlabel('x values');
plt.ylabel('y values');
plt.title('y vs x values');

# scatterplot subplot
# plt.subplot(2,2,3)
# plt.text(-0.5, 0, 'Hey, look at my figure!')
```
