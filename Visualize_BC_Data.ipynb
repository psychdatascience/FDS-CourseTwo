# Visualize Breast Cancer Data with seaborn

## Import libraries

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

## Load data

```python
bc_df = pd.read_csv('FP_breast_cancer_data_catcol.csv')
```

```python
bc_df
```

```python
bc_df.columns
```

## Look at pair plots of some variables

```python
sub_cols = ['mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity']
sns.pairplot(bc_df, hue = 'target_category', vars = sub_cols)
```

```python
sub_cols = ['mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error']
sns.pairplot(bc_df, hue = 'target_category', vars = sub_cols)
```

---

It looks like "mean concave points" and "mean perimeter" do pretty well. Perhaps we should use them!

```python
sub_cols = ['mean concave points', 'mean perimeter']
sns.pairplot(bc_df, hue = 'target_category', vars = sub_cols)
```
