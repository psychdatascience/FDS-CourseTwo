# matplotlib homework

1. Make a plot of a straight line. Use `linspace()` to create the x values and y = a + b*x to create the y values (use an a and b of your choosing).

2. Make some data that are straight line values from the same relationship as in 1. plus random noise. Plot these 

3. Plot the straight line from 1. and the data from 2. on the same graph. Make sure to add a legend.

```python
import numpy as np
import matplotlib.pyplot as plt

a , b = 5, 1
my_x = np.linspace(-3, 3)
my_y = a + b * my_x

my_x_dat = np.linspace(-2, 2, 25)
my_y_dat = a + b * my_x_dat + np.random.randn(25)

plt.plot(my_x, my_y, 'b--', label = 'trend')
plt.plot(my_x_dat, my_y_dat, 'ob', label = 'data')
plt.legend()
```
