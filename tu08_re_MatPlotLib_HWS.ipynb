1. Make a plot of a straight line. Use `linspace()` to create the x values and y = a + b*x to create the y values (use an a and b of your choosing).

```python
my_x = np.linspace(0, 10, 50)

a, b = 0, 2.2
my_y = a + b * my_x

plt.plot(my_x, my_y)
```

2. Make some data that are straight line values from the same relationship as in 1. plus random noise. Plot these 

```python
dat_x = np.linspace(0, 10, 10)

a, b = 0, 2.2
noys_amp = 0.3
dat_y = a + b * dat_x + np.random.randn(len(dat_x))

plt.plot(dat_x, dat_y, 'o')
```

3. Plot the straight line from 1. and the data from 2. on the same graph. Make sure to add the standard annotations, including a legend.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(my_x, my_y, 'b--', label = 'trend')
plt.plot(dat_x, dat_y, 'ob', label = 'data')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('data and line')
plt.legend()
```

5. Make 500 ***normally*** distributed random numbers and make a histogram of them.

```python
x1 = np.random.randn(500)
plt.hist(x1, bins = 40, alpha = 0.2, label = 'Normal');
```

6. Make 500 ***uniformly*** distributed random numbers (use `...rand()` instead of `...randn()`) and make a histogram of them.

```python
x2 = np.random.rand(500)
plt.hist(x2, bins = 40, color = 'g', alpha = 0.2, label = 'Uniform');
```
