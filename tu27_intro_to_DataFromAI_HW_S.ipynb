## Part 3. Plan the measurement strategy

Instead of assigning just one emotion category, you will ask the model to rate **four emotional dimensions** for each response:

- `anxiety`
- `joy`
- `frustration`
- `calm`

Each dimension should receive a score from **1 to 5**.

This is an important design choice. A multi-dimensional rating system can capture more nuance than a single category label. In particular, it may reveal whether mixed-emotion responses are rated highly on more than one dimension.

Before coding the whole dataset, you will begin by writing a function that rates one response at a time.

### Task 1. Write a reusable function

Write a function called `rate_multiple_emotions()` that takes one text response as input and returns a Python dictionary with these four keys:

- `anxiety`
- `joy`
- `frustration`
- `calm`

Your prompt should instruct the model to return **valid JSON only**, with integer values from 1 to 5.

Include simple error handling:
- if JSON parsing fails, print the raw output
- try to recover the JSON by extracting the text between the first `{` and the last `}`
- if recovery fails, return `None` values

This function is the core measurement tool for the assignment.

```python
def rate_multiple_emotions(text, model="gpt-4.1-mini"):
    prompt = f'''
    Read the sentence below and rate the extent to which it expresses each of the following emotions.

    Sentence: "{text}"

    Return your answer in valid JSON with exactly these keys:
    - "anxiety"
    - "joy"
    - "frustration"
    - "calm"

    Rules:
    - Each value should be an integer from 1 to 5.
    - 1 means the emotion is barely present.
    - 5 means the emotion is strongly present.
    - Return JSON only. Do not include any extra text before or after the JSON object.
    '''

    output_text = None

    try:
        response = client.responses.create(
            model=model,
            input=prompt
        )
        output_text = response.output_text.strip()
        return json.loads(output_text)

    except json.JSONDecodeError:
        print("Warning: JSON parsing failed. Raw output was:")
        print(output_text)
        try:
            start = output_text.find("{")
            end = output_text.rfind("}") + 1
            cleaned_text = output_text[start:end]
            return json.loads(cleaned_text)
        except Exception:
            print("Warning: Could not recover valid JSON.")
            return {"anxiety": None, "joy": None, "frustration": None, "calm": None}

    except Exception as e:
        print("Warning: API call failed.")
        print(e)
        time.sleep(2)
        return {"anxiety": None, "joy": None, "frustration": None, "calm": None}
```

## Part 4. Run the experiment across both conditions

Now you are ready to apply your function to all responses.

Create a list called `all_results`. For each response, store:

- the original text
- the condition (`"single"` or `"mixed"`)
- the four emotion ratings

Use a loop to process the single-emotion responses and the mixed-emotion responses. Add a short pause such as `time.sleep(1)` between API calls to reduce the chance of rate-limit errors.

The goal of this section is to build a dataset that combines both conditions in one place.

```python
all_results = []

for text in single_emotion_responses:
    ratings = rate_multiple_emotions(text)
    ratings["response_text"] = text
    ratings["condition"] = "single"
    all_results.append(ratings)
    time.sleep(1)

for text in mixed_emotion_responses:
    ratings = rate_multiple_emotions(text)
    ratings["response_text"] = text
    ratings["condition"] = "mixed"
    all_results.append(ratings)
    time.sleep(1)

all_results[:3]
```

## Part 6. Compare average emotion scores by condition

Now that your dataset is ready, you can analyze the experiment.

Your first goal is to compare the **average emotion ratings** for the two conditions.

Group the data by `condition` and compute the mean for these four variables:

- `anxiety`
- `joy`
- `frustration`
- `calm`

This will tell you whether the mixed-emotion responses tend to produce a different overall pattern from the single-emotion responses.

```python
condition_means = df.groupby("condition")[["anxiety", "joy", "frustration", "calm"]].mean()
condition_means
```

Create a plot that compares the average emotion scores for the two conditions.

## Part 8. Analyze emotional complexity more directly

The grouped averages are useful, but they do not fully capture the idea of **mixedness**.

In this section, create a new variable called `high_emotion_count` that counts how many of the four emotion dimensions have a score of **4 or higher** for each response.

For example:
- if one response has scores `[5, 1, 1, 1]`, the count is 1
- if another response has scores `[4, 4, 1, 2]`, the count is 2

This gives you a simple way to quantify whether a response shows one dominant emotion or multiple strong emotions.

After creating this new variable, compare its average value across the two conditions.

```python
df["high_emotion_count"] = (
    (df["anxiety"] >= 4).astype(int) +
    (df["joy"] >= 4).astype(int) +
    (df["frustration"] >= 4).astype(int) +
    (df["calm"] >= 4).astype(int)
)

high_count_means = df.groupby("condition")["high_emotion_count"].mean()
high_count_means
```

### Optional visualization for emotional complexity

Create a simple bar plot of the average `high_emotion_count` by condition.

This provides a compact visual answer to the question:  
**Do mixed-emotion responses show more highly rated emotional dimensions than single-emotion responses?**

```python
high_count_means.plot(kind="bar")
plt.xlabel("Condition")
plt.ylabel("Average Number of High Emotion Scores")
plt.title("Average Emotional Complexity by Condition")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
